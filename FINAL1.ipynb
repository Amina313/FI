## 4. Разделение данных

Разделите данные на обучающую и тестовую выборки (например, 80% на обучение и 20% на тестирование).

```python
X = home_data[['status', 'propertyType', 'baths', 'fireplace', 'city', 'sqft', 'zipcode', 'beds', 'state', 'stories', 'PrivatePool']]
y = home_data[['target']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

## 5. Обучение моделей

Обучите выбранные модели на обучающей выборке.

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50);
```

## 6. Оценка моделей

Используйте метрики для оценки качества моделей, такие как:


Средняя абсолютная ошибка (MAE)
Среднеквадратичная ошибка (MSE)
R² (коэффициент детерминации)

```python
mae = metrics.mean_absolute_error(y_test, predictions)
mse = metrics.mean_squared_error(y_test, predictions)
r2 = metrics.r2_score(y_test, predictions)

print(f'MAE: {mae}, MSE: {mse}, R²: {r2}')

print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
